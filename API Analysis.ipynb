### Number of Authors and their Articles

```python
import requests
import datetime
import dateutil.parser
from operator import itemgetter
```

```python
URL = 'http://mertium.com/api/news/trending/articles/last/week/'
r = requests.get(url = URL)
data = r.json()

articles = []
repeated_articles = []
titles_set = set()
page = 1
while True:
    
    for index in range(len(data['results'])):
        
        article = {}
        article['title'] = data['results'][index]['title']
        article['news_source'] = data['results'][index]['news_source']
        time = dateutil.parser.parse(data['results'][index]['timestamp'])
        article['timestamp'] = time
        article['link'] = data['results'][index]['link']
        article['twitter_score'] = data['results'][index]['tweet_count']
        article['articletags_set'] = data['results'][index]['articletags_set']
        article['articletags_set'] = data['results'][index]['articletags_set']
        article['article_id'] = data['results'][index]['article_id']
        try:
            authorName = data['results'][index]['authors'][0]['name']
            if authorName == '':
                article['author'] = 'Empty'
            else:
                article['author'] = authorName.strip()
        except:
            article['author'] = 'Empty'
        article['timestamp_clean'] = time.strftime("%m/%d/%Y, %H:%M:%S")
        article['link'] = data['results'][index]['link']

        if article['article_id'] in titles_set:
            repeated_articles.append(article['article_id'])
        else:
            articles.append(article)
        titles_set.add(article['article_id'])
        
    URL = data['next']
    r = requests.get(url = URL)
    data = r.json()
    
    if data['next'] == None:
        
        for index in range(len(data['results'])):
            
            article = {}
            article['title'] = data['results'][index]['title']
            article['news_source'] = data['results'][index]['news_source']
            time = dateutil.parser.parse(data['results'][index]['timestamp'])
            article['timestamp'] = time
            article['link'] = data['results'][index]['link']
            article['twitter_score'] = data['results'][index]['tweet_count']
            article['articletags_set'] = data['results'][index]['articletags_set']
            article['articletags_set'] = data['results'][index]['articletags_set']
            article['article_id'] = data['results'][index]['article_id']
            
            try:
                authorName = data['results'][index]['authors'][0]['name']
                if authorName == '':
                    article['author'] = 'Empty'
                else:
                    article['author'] = authorName.strip()
            except:
                article['author'] = 'Empty'
            article['timestamp_clean'] = time.strftime("%m/%d/%Y, %H:%M:%S")
            article['link'] = data['results'][index]['link']

            if article['article_id'] in titles_set:
                repeated_articles.append(article['article_id'])
            else:
                articles.append(article)
            titles_set.add(article['article_id'])
            
        break
```

### Number of Articles in 24 Hrs

```python
allArticlesSet = set()
now = datetime.datetime.now(dateutil.tz.tzoffset(None, -18000))
one_day_before = now - datetime.timedelta(hours=24)
print(now, one_day_before)
def getUniqueArticles(data, hours):

    now = datetime.datetime.now(dateutil.tz.tzoffset(None, -18000))
    one_day_before = now - datetime.timedelta(hours=hours)

    articles = set()
    while True:

        for index in range(len(data['results'])):

            article = {}
            headline = data['results'][index]['title']
            headline = headline.replace('&#39;', '\'')
            article['title'] = headline
            article['news_source'] = data['results'][index]['news_source']
            time = dateutil.parser.parse(data['results'][index]['timestamp'])
            article['timestamp'] = time
            article['link'] = data['results'][index]['link']
            article['twitter_score'] = data['results'][index]['tweet_count']
            article['articletags_set'] = data['results'][index]['articletags_set']

            if (time >= one_day_before) and (article['title'] not in allArticlesSet):
                print(article['twitter_score'])
                articles.add(article['title'])
                allArticlesSet.add(article['title'])

        URL = data['next']
        r = requests.get(url = URL)
        data = r.json()

        if data['next'] == None:
            for index in range(len(data['results'])):

                article = {}
                headline = data['results'][index]['title']
                headline = headline.replace('&#39;', '\'')
                article['title'] = headline
                article['news_source'] = data['results'][index]['news_source']
                time = dateutil.parser.parse(data['results'][index]['timestamp'])
                article['timestamp'] = time
                article['link'] = data['results'][index]['link']
                article['twitter_score'] = data['results'][index]['tweet_count']
                article['articletags_set'] = data['results'][index]['articletags_set']

                if (time >= one_day_before) and (article['title'] not in allArticlesSet):
                    print(article['twitter_score'])
                    articles.add(article['title'])
                    allArticlesSet.add(article['title'])


            break
#         print(data['next'])
    return len(articles)
```

```python
URL = 'http://mertium.com/api/news/trending/articles/last/week/'
r = requests.get(url = URL)
data = r.json()

weeklyNumber = []
for hrs in [24]: #,48,72,96,120,144,168]:
    ans = getUniqueArticles(data, hrs)
    weeklyNumber.append(ans)
    print('Last {} hours ago: '.format(hrs), ans)
```
